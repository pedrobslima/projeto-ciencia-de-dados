**Dicionário do dataset:**

(Disponível em: https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/Febre+Amarela/Dicionario+de+variaveis.pdf)

* ID: Identificador sequencial único
* MACRORREG_OCOR: Sigla da macrorregião do local de ocorrência
* COD_UF_OCOR: Código IBGE da Unidade Federada do local de ocorrência
* UF_OCOR: Sigla da Unidade Federada do local de ocorrência
* COD_MUN_OCOR: Código IBGE do município do local de ocorrência
* MUN_OCOR: Nome do município do local de ocorrência
* DATA_OCOR: Data de ocorrência do evento (dd/mm/aaaa)
* SE_OCOR: Semana epidemiológica de ocorrência do evento
* MES_OCOR: Mês de ocorrência do evento
* ANO_OCOR: Ano de ocorrência do evento
* MONITORAMENTO_OCOR: Período de monitoramento de ocorrência do evento¹

¹: o **período de monitoramento** corresponde à estratificação temporal dos dados em períodos
anuais com início em julho e término em junho. Cada período de monitoramento corresponde
a um intervalo de 12 meses, que inclui o segundo semestre de um ano e o primeiro semestre do
ano seguinte. Essa representação do componente temporal decorre do reconhecimento de um
período sazonal de transmissão, entre dezembro e maio, que concentra a maior parte dos
eventos registrados no país, e tem como intuito evitar a análise fragmentada dos processos de
transmissão, cujo pico de ocorrência se dá geralmente na transição entre os anos. O Ministério
da Saúde adota essa estratificação para fins de políticas de vigilância em saúde e análise
epidemiológica.

**Dicionário de dados:** (ESPÉCIME)

* SEQ_TAD - Chave que identifica o termo de apreensão;
* SEQ_ESPECIME_APREENDIDO - Chave que identifica o espécime apreendido;

* DES_UNIDADE_MEDIDA - Unidade de medida da quantidade de espécimes apreendidos.

* GRUPO - Identifica o grupo de organismos dos espécimes apreendidos: Anfíbio; Ave; Caranguejo; Domestico; Exótico; Invertebrado; Lagosta; Mamífero; Molusco; Outros; Palmito; Peixes Ornamentais; Pescado; Réptil;

* NOME_CIENTIFICO - Nome científico do espécime apreendido;
* DES_NOME_POPULAR - Nome popular do espécime apreendido;
* CITES - Identifica se a espécie do espécime apreendido consta na lista de espécies Cites (Convention on International Trade in Endangered Species of Wild Fauna and Flora): S – Sim; N – Não;
* OBSERVACAO - Descreve demais circunstâncias da apreensão e da situação do espécime;
* DAT_ALTERACAO - Data da última alteração do registro de espécime (ISO 8601);
* ULTIMA_ATUALIZACAO_RELATORIO- Data da última atualização do recurso no portal de dados abertos.

* CPF_CNPJ_PESSOA_APREENSAO - Número de Cadastro Nacional de Pessoa Jurídica (CNPJ) ou Cadastro de Pessoa Física (CPF) da pessoa responsável pelo bem apreendido;
* NUM_PROCESSO - Número do processo administrativo do Ibama em que será processada a apreensão;
* DES_TAD - Descrição das circunstâncias da apreensão, justificativa e normas que fundamentaram a aplicação da medida;
* COD_MUNICIPIO - Código do município (IBGE) referente ao local da apreensão;
* NOM_MUNICIPIO - Nome do município referente ao local da apreensão;
* SIG_UF - Sigla da unidade da federação referente ao local da apreensão;
* DES_LOCALIZACAO - Descrição do local da apreensão;
* NUM_LONGITUDE_TAD - Longitude do local da apreensão;
* NUM_LATITUDE_TAD - Latitude do local da apreensão;
* FORMA_ENTREGA - Meio empregado para dar ciência da apreensão à pessoa responsável pelo bem apreendido: pessoalmente; representante; aviso de recebimento (AR); edital; o autuado recusou-se a assinar;
* UNID_APRESENTACAO - Unidade responsável pelo recebimento e instauração do processo;
* UNID_CONTROLE - Unidade responsável pelo gerenciamento do termo no sistema de fiscalização;
* VALOR_TAD - Somatória do valor de todos os bens apreendidos identificados no termo;
* SEQ_AUTO_INFRACAO - Chave que identifica o auto de infração vinculado ao termo de apreensão;
* SEQ_NOTIFICACAO - Chave que identifica a notificação vinculada ao termo de apreensão;
* SEQ_ACAO_FISCALIZATORIA - Chave que identifica a ação fiscalizatória em que ocorreu a apreensão;
* NOM_OPERACAO - Nome da operação de fiscalização;
* SEQ_ORDEM_FISCALIZACAO - Chave que identifica a ordem de fiscalização que motivou a ação fiscalizatória;
* NUM_ORDEM_FISCALIZACAO - Identificação da ordem de fiscalização que motivou a ação fiscalizatória, com número de série e estado de origem da unidade que emitiu a ordem;
* SEQ_SOLICITACAO_RECURSO - Chave que identifica a solicitação de recurso vinculada à ordem de fiscalização;
* NUM_SOLICITACAO_RECURSO - Número da solicitação de recurso vinculada à ordem de fiscalização;
* OPERACAO_SOL_RECURSO - Nome da operação de fiscalização descrita na solicitação de recurso;
* DAT_ALTERACAO - Data da última alteração registrada para o termo (ISO 8601);
