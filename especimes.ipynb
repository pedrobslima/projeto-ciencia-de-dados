### **Dicionário de dados:** (TERMOS DE APREENSÃO)

* SEQ_TAD - Chave que identifica o termo de fiscalização;
* STATUS_FORMULARIO - Situação atual do formulário:
  * Lavrado;
  * Cancelado;
  * Substituído por outro;
  * Excluído;
* SIT_CANCELADO - Indica se o termo encontra-se cancelado:
  * S (Sim);
  * N (Não);
* NUM_TAD - Número impresso no formulário que compõe a identificação do termo juntamente com a série;
* SER_TAD - Caractere impresso no formulário que compõe a identificação do termo juntamente com o número;
* DAT_TAD - Data em que o termo foi lavrado (ISO 8601);
* DAT_IMPRESSAO - Data de impressão do termo no aparelho PDA (ISO 8601);
* NUM_PESSOA_APREENSAO - Número de registro da pessoa no banco de dados do Ibama;
* NOM_PESSOA_APREENSAO - Nome da pessoa responsável pelo bem apreendido;
* CPF_CNPJ_PESSOA_APREENSAO - Número de Cadastro Nacional de Pessoa Jurídica (CNPJ) ou Cadastro de Pessoa Física (CPF) da pessoa responsável pelo bem apreendido;
* NUM_PROCESSO - Número do processo administrativo do Ibama em que será processada a apreensão;
* DES_TAD - Descrição das circunstâncias da apreensão, justificativa e normas que fundamentaram a aplicação da medida;
* COD_MUNICIPIO - Código do município (IBGE) referente ao local da apreensão;
* NOM_MUNICIPIO - Nome do município referente ao local da apreensão;
* SIG_UF - Sigla da unidade da federação referente ao local da apreensão;
* DES_LOCALIZACAO - Descrição do local da apreensão;
* NUM_LONGITUDE_TAD - Longitude do local da apreensão;
* NUM_LATITUDE_TAD - Latitude do local da apreensão;
* FORMA_ENTREGA - Meio empregado para dar ciência da apreensão à pessoa responsável pelo bem apreendido:
  * pessoalmente;
  * representante;
  * aviso de recebimento (AR);
  * edital;
  * o autuado recusou-se a assinar;
* UNID_APRESENTACAO - Unidade responsável pelo recebimento e instauração do processo;
* UNID_CONTROLE - Unidade responsável pelo gerenciamento do termo no sistema de fiscalização;
* VALOR_TAD - Somatória do valor de todos os bens apreendidos identificados no termo;
* SEQ_AUTO_INFRACAO - Chave que identifica o auto de infração vinculado ao termo de apreensão;
* SEQ_NOTIFICACAO - Chave que identifica a notificação vinculada ao termo de apreensão;
* SEQ_ACAO_FISCALIZATORIA - Chave que identifica a ação fiscalizatória em que ocorreu a apreensão;
* NOM_OPERACAO - Nome da operação de fiscalização;
* SEQ_ORDEM_FISCALIZACAO - Chave que identifica a ordem de fiscalização que motivou a ação fiscalizatória;
* NUM_ORDEM_FISCALIZACAO - Identificação da ordem de fiscalização que motivou a ação fiscalizatória, com número de série e estado de origem da unidade que emitiu a ordem;
* SEQ_SOLICITACAO_RECURSO - Chave que identifica a solicitação de recurso vinculada à ordem de fiscalização;
* NUM_SOLICITACAO_RECURSO - Número da solicitação de recurso vinculada à ordem de fiscalização;
* OPERACAO_SOL_RECURSO - Nome da operação de fiscalização descrita na solicitação de recurso;
* DAT_ALTERACAO - Data da última alteração registrada para o termo (ISO 8601);
* TIPO_ALTERACAO - Tipo da última alteração registrada no termo:
  * Redistribuído;
  * Preenchido;
  * Alterado;
  * Cancelado;
  * Cancelamento estornado;
* JUSTIFICATIVA_ALTERACAO - Descrição do motivo da alteração realizada no termo;
* ULTIMA_ATUALIZACAO_RELATORIO - Data da última atualização do recurso no portal de dados abertos.

````python
mask1 = df['QTD_ESPECIME_APREENDIDO'].isna()
mask2 = df['DES_UNIDADE_MEDIDA'].isna()
mask3 = (mask1 | mask2) & ~(mask1 & mask2) # xor
mask4 = ~(mask1) & mask2 # se a qtd não é NAN, mas a unidade sim
mask5 = mask1 & ~(mask2) # se qtd é NAN, mas a unidade não
````

````python
print(f'''xor:\t\t\t{len(df[mask3])}
QTD!=NA & UND=NA:\t{len(df[mask4])}
QTD=NA & UND!=NA:\t{len(df[mask5])}''')
````

No caso da ```mask4```, dá para tentar dar um palpite adequado sobre a unidade de medida que está faltando vendo qual o espécime, tipo e, possivelmente, o valor da quantidade

````python
df[mask4].loc[:, 'TIPO'].value_counts()
````

````python
mask6 = ~(mask1) & ~(mask2)
````

````python
# transformando em floats, menos os <NAN>, dps lido com eles
df["QTD_ESPECIME_APREENDIDO"] = df["QTD_ESPECIME_APREENDIDO"].apply(lambda x: float(x.replace(',', '.')) if type(x)==str else np.NAN)
````

````python
cols = ['QTD_ESPECIME_APREENDIDO', 'DES_UNIDADE_MEDIDA']#, 'GRUPO', 'SITUACAO_CARACTERISTICA']
df[(df['TIPO'] == 'Fauna') & (mask6)].loc[:, cols].groupby(by='DES_UNIDADE_MEDIDA').aggregate('median')#.head()
````
